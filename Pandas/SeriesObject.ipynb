# Pandas Series Object

##### series object in pandas is a one-dimensional labeled array capable of holding data of any type (integer, string, float, python objects, etc.). The axis labels are collectively referred to as the index. A pandas Series can be created using various inputs such as lists, dictionaries, or scalar values.


### Series object from List

```python
import pandas as pd
```

```python
s1 = pd.Series([10, 20, 30, 40, 50])
print(s1)
print(type(s1))
```

#### Changing Index

```python
s1.index = ['a', 'b', 'c', 'd', 'e']
print(s1)
```

```python
s1 = pd.Series([10, 20, 30, 40, 50],
               index=['a', 'b', 'c', 'd', 'e'])
print(s1)
```

### Series Object from dictionary

```python
s1 = pd.Series({'a': 10, 'b': 20, 'c': 30, 'd': 40, 'e': 50})
print(s1)
print(type(s1))
```

#### changing Indices

```python
s1.index = ['1', '2', '3', '4', '5']
print(s1)
```

```python
s1 = pd.Series({'1': 'Delhi',
                '2': 'Washington DC',
                '3': 'London',
                '4': 'Paris',},
               index=['3', '4', 'UK', 'Germany'])
print(s1)
# we only can change the orders of existing indices not
#but using s1.index we can change to any values
```

#### Extracting Individual Element

```python
s1 = pd.Series([10, 20, 30, 40, 50])
print(s1[0])
print("from end \n" ,s1[-3:]) # from end
print("slicing\n ",s1[:3]) # slicing or extracting sequence of elements
```

### Basic maths operations on Series

##### Adding a Scalar Value to elements

```python
s1+= 5
print(s1)
```

##### Adding two Series Objects

```python
s1 = pd.Series([10, 20, 30, 40, 50])
s2 = pd.Series([5, 15, 25, 35, 45])
s3 = s1 + s2
print(s3)
```
