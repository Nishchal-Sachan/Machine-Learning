## More Pandas Functions 

```python
import pandas as pd
```

```python
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
iris.mean(numeric_only=True)
## Calculate the mean of each numeric column
```

```python
iris.median(numeric_only=True)
## Calculate the median of each numeric column
```

```python
iris.min(numeric_only=True)
## Calculate the minimum value of each numeric column
```

```python
iris.max(numeric_only=True)
## Calculate the maximum value of each numeric column
```

### Applying User Defined Functions

```python
iris.head()
```

```python
def double_maker(x):
    return x * 2
```

```python
iris[['Sepal.Width','Petal.Width']].apply(double_maker)
```

```python
iris['Species'].value_counts()
## Count the occurrences of each unique value in the 'Species' column
```

```python
iris.sort_values(by='Sepal.Length')
## Sort the DataFrame by the 'Sepal.Length' column in ascending order
```
